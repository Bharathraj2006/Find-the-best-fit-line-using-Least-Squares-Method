```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.array(eval(input()))
y = np.array(eval(input()))
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
num=0
denom = 0
for i in range(len(x)):
    num += (x[i] - x_mean)*(y[i] - y_mean)
    denom += (x[i] - x_mean)**2
```

```python
m = num/denom
c = y_mean - m*x_mean
print(m,c)
```

```python
y_predicted = m*x + c
print(y_predicted)
```

```python
plt.scatter(x,y)
plt.plot(x,y_predicted,color='red')
plt.show()
```
